```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Amit', 'Neha', 'Raj', 'Priya', 'Karan', 'Sneha'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Age': [21, 22, 20, 23, 21, 22],
    'Income': [25000, 30000, 22000, 35000, 28000, 32000]
}

df = pd.DataFrame(data)

print(df)
```

Calculate Summary Statistics Grouped by Gender

```python
grouped = df.groupby('Gender')['Income']

print("Mean")
print(grouped.mean())

print("\nMedian")
print(grouped.median())

print("\nMinimum")
print(grouped.min())

print("\nMaximum")
print(grouped.max())

print("\nStandard Deviation")
print(grouped.std())
```

Create List Containing Numeric Values for Each Category

```python
income_list = df.groupby('Gender')['Income'].apply(list)

print(income_list)
```
